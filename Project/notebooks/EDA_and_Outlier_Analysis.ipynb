# Exploratory Data Analysis (EDA) & Outlier Analysis

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- Cell 1: Load the Cleaned Data ---
# Define the full path to your cleaned CSV file.
cleaned_file_path = r"C:\Users\DELL\bootcamp_Dhriti_Trivedi\Project\Data\processed\yahoo_data_cleaned.csv"

# Load the preprocessed data from the specific path.
# index_col='Date' and parse_dates=True ensure the 'Date' column is treated as the index.
try:
    df = pd.read_csv(cleaned_file_path, index_col='Date', parse_dates=True)
    print("Cleaned dataset loaded successfully!")
    print(df.head())
except FileNotFoundError:
    print(f"Error: The file was not found at the path: {cleaned_file_path}")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
#--- Cell 2: Descriptive Statistics ---
# Get a statistical summary of the main numerical columns.
print("\n--- Descriptive Statistics ---")
print(df[['Close', 'Volume']].describe())
```

```python
# --- Cell 3: Visualizing Distributions ---
# Plot histograms to understand the distribution of the 'Close' price and 'Volume'.
print("\n--- Plotting Distributions ---")
plt.style.use('seaborn-v0_8-whitegrid')
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))

# Close Price Distribution
sns.histplot(df['Close'], kde=True, ax=axes[0], bins=50)
axes[0].set_title('Distribution of Closing Price')
axes[0].set_xlabel('Closing Price')

# Volume Distribution
sns.histplot(df['Volume'], kde=True, ax=axes[1], color='orange', bins=50)
axes[1].set_title('Distribution of Trading Volume')
axes[1].set_xlabel('Volume')

plt.tight_layout()
plt.show()
```

## Closing Price Over Time
This line graph illustrates the daily closing price of the asset over five years. It reveals several key trends:

- A period of growth from 2018 to the beginning of 2020.

- A sharp decline around March 2020, which corresponds to the economic impact of the COVID-19 pandemic.

- A strong recovery and subsequent bull run through 2020 and 2021.

- Increased volatility and a general downturn in 2022, followed by a partial recovery into 2023.

- This chart provides a high-level view of the market's performance and captures major economic events through its price fluctuations.

## Trading Volume Over Time

This graph shows the number of shares traded each day. Unlike the price, the volume doesn't show a consistent long-term trend but rather highlights periods of high or low market activity. Key observations include:

- Spikes in volume often coincide with periods of high price volatility or major market events, such as the sharp decline in 2020. High volume during a large price drop can indicate panic selling.

- Periods of lower volume might occur during stable or sideways market trends.

```python
# --- Cell 4: Outlier Analysis using Box Plots ---
# Use box plots to visually inspect for outliers in price and volume.
print("\n--- Outlier Analysis ---")
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))

# Close Price Box Plot
sns.boxplot(y=df['Close'], ax=axes[0])
axes[0].set_title('Box Plot of Closing Price')
axes[0].set_ylabel('Closing Price')

# Volume Box Plot
sns.boxplot(y=df['Volume'], ax=axes[1], color='orange')
axes[1].set_title('Box Plot of Trading Volume')
axes[1].set_ylabel('Volume')

plt.tight_layout()
plt.show()
```
