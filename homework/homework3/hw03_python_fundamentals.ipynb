```python
import numpy as np

# Create a NumPy array
prices = np.array([100, 105, 110, 120, 125])

# Perform elementwise operations
percent_changes = (prices[1:] - prices[:-1]) / prices[:-1] * 100
print("Percent Changes:", percent_changes)

```
